```python
import matplotlib.pyplot as plt
import numpy as np
```

# Types of Graphs
1. plot -       line graph ,between two parameters
2. bar -        bar graph of single parameter
3. hist -       histogram of a bunch of frequency of single parameter
4. scatter-     dot graph between two parameters
5. boxplot/box- statistical overview in form of box for a single parameter mean,outliers,quantiles(25%, 75%)(optional -categorical data) 
6. stem -       more like bar graph ,lines instead of bars
7. violinplot-  violin shaped statistical overview graph 
                spread of data (more like hist)
                single parameter(optional- cat-data)

```python
x=np.linspace(0,5,11)
x1=np.arange(1,11,3)
```

```python
x
```

```python
x1
```

```python
y=x**3
```

```python
plt.plot(x,y)
plt.title('plot')
plt.show()
```

```python
plt.bar(x,y)
plt.title('bar graph')
plt.show()
```

```python
plt.hist(x)
plt.title('histogram')
plt.show()
```

```python
plt.scatter(x,y)
plt.title('scatter')
plt.grid()
plt.xlabel('x--')
plt.ylabel('y--')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.boxplot(x)
plt.title('box plot')
plt.show()
```

```python
plt.stem(x,y)
plt.title('stem')
plt.show()
```

```python
plt.violinplot(x)
plt.title('Violen Plot')
plt.show()
```

```python
y2=x**4
```

```python
plt.plot(x,y,x,y2)
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,lw=3,marker='o',markerfacecolor='yellow',markersize='4')
plt.plot(x,y2,marker='v',markersize='10',markerfacecolor='green')
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```
